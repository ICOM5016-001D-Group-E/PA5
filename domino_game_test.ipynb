Imports

```python
from game.game_state import GameState
from agents.random_agent import RandomAgent
from agents.heuristic_agent import HeuristicAgent
from agents.smart_heuristic_agent import SmartHeuristicAgent
from agents.monte_carlo_agent import MonteCarloAgent
```

Game Runner

```python
def play_ai_game(agent0, agent1, agent2, agent3):
    game = GameState()
    game.setup()

    agents = [agent0, agent1, agent2, agent3]

    while not game.game_over:
        player = game.current_player
        agent = agents[player]

        # handle forced first move
        if game.board.is_empty() and game.forced_first_tile is not None:
            tile = game.forced_first_tile
            side = None
        else:
            tile, side = agent.choose_move(game)

        if tile is None:
            game.apply_pass()
        else:
            game.apply_move(tile, side)

    return game.winner, game.scores
```

Single Test

```python
winner, scores = play_ai_game(
    RandomAgent(),
    HeuristicAgent(),
    SmartHeuristicAgent(),
    MonteCarloAgent()
)

print("Winner:", winner)
print("Scores:", scores)
```

Run Multiple Games 

```python
def run_experiment(agentA, agentB, games=100):
    wins = {0: 0, 1: 0, None: 0}

    for _ in range(games):
        winner, _ = play_ai_game(
            agentA(), agentB(), agentA(), agentB()
        )
        wins[winner] += 1

    return wins

def win_rate(results):
    total = sum(results.values())
    return {k: round(v / total, 3) for k, v in results.items()}
```

Experiments

```python
print("Random vs Random")
res = run_experiment(RandomAgent, RandomAgent, 200)
print(res)
print("Win rate:", win_rate(res))

print("\nHeuristic vs Random")
res = run_experiment(HeuristicAgent, RandomAgent, 200)
print(res)
print("Win rate:", win_rate(res))

print("\nSmart Heuristic vs Random")
res = run_experiment(SmartHeuristicAgent, RandomAgent, 200)
print(res)
print("Win rate:", win_rate(res))
```

Monte Carlo Evaluation

```python
print("\nMonte Carlo vs Random")
res = run_experiment(MonteCarloAgent, RandomAgent, 200)
print(res)
print("Win rate:", win_rate(res))

print("\nMonte Carlo vs Heuristic")
res = run_experiment(MonteCarloAgent, HeuristicAgent, 200)
print(res)
print("Win rate:", win_rate(res))

print("\nMonte Carlo vs Smart Heuristic")
res = run_experiment(MonteCarloAgent, SmartHeuristicAgent, 200)
print(res)
print("Win rate:", win_rate(res))
```

Strong Comparison

```python
GAMES = 500  
print(f"\nFINAL TEST ({GAMES} games)\n")

print("Monte Carlo vs Random")
res = run_experiment(MonteCarloAgent, RandomAgent, GAMES)
print(res)
print("Win rate:", win_rate(res))

print("\nMonte Carlo vs Heuristic")
res = run_experiment(MonteCarloAgent, HeuristicAgent, GAMES)
print(res)
print("Win rate:", win_rate(res))

print("\nMonte Carlo vs Smart Heuristic")
res = run_experiment(MonteCarloAgent, SmartHeuristicAgent, GAMES)
print(res)
print("Win rate:", win_rate(res))
```
